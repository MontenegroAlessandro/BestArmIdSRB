```python
# Libraries
import math
import matplotlib.pyplot as plt
import random
```

```python
# Define the horizon
T = 500
eps = 1/3
sigma = .1
T_plot = 500
```

```python
f1 = lambda x : .0025*x if .0025*x < 1 else 1
f2 = lambda x : .47 * (1 - 10/(10+x))
f3 = lambda x : .004*x if .004*x < .7 else .7
```

```python
arms = [f1, f2, f3]
convergence = [.47, 1, .7]
```

```python
# Plot the Arms
plt.clf()

i = 1
for arm in arms:
    label = str(i)
    plt.plot([arm(j) for j in range(T_plot)], label=label)
    i += 1

plt.grid()
plt.legend(loc="best")
```
