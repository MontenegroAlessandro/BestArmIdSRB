```python
# Libraries
import math
import matplotlib.pyplot as plt
import random
import tikzplotlib as tkz
```

```python
# Define the horizon
T = 500
eps = 1/3
sigma = .1
T_plot = 500
```

```python
f1 = lambda x : .0025*x if .0025*x < 0.99 else 0.99
f2 = lambda x : .47 * (1 - 10/(10+x))
f3 = lambda x : .004*x if .004*x < .7 else .7
```

```python
arms = [f1, f2, f3]
convergence = [1, .47, .7]
```

```python
# Plot the Arms
plt.clf()

i = 1
for arm in arms:
    label = str(i)
    plt.plot([arm(j) for j in range(T_plot)], label=label)
    i += 1

plt.grid()
plt.legend(loc="best")
# tkz.save("exp2_arms.tex")
```

```python
df1 = lambda x: 0.0025 if 0.0025*x < 0.99 else 0
df2 = lambda x: (0.47*10)/((10+x)**2)
df3 = lambda x: 0.004 if 0.004*x < 0.7 else 0

beta = 1.001
beta_fun = lambda x: math.pow(x, -beta)

darms = [df1, df2, df3, beta_fun]

# Plot the Arms
plt.clf()

i = 1
for arm in darms:
    label = str(i)
    plt.plot([arm(j) for j in range(200, T_plot)], label=label)
    i += 1

plt.grid()
plt.legend(loc="best")
```
