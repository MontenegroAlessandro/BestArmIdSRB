```python
# Libraries
# import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
T = 100
```

```python
fun = lambda x, a, b, c : a * math.pow(x / b, c)
def draw(ax, bx, cx):
    lab = str(ax) + ', ' + str(bx) + ', ' + str(cx)
    plt.plot([fun(i, ax, bx, cx) for i in range(T)], label=lab)
    plt.grid()
    plt.legend(loc='best')
    #plt.clf()
```

```python
draw(0.24, 1, 1/2)
draw(0.4, 1, 1/3)
```

```python
draw(0.3, 7, 1/4)
draw(0.3, 7, 1/3)
draw(0.23, 7, 1/2)
draw(0.23, 10, 1/3)
draw(0.24, 6, 1/2)
```

```python
foo = lambda x, a, c : a * math.pow(x, c)
def draw_foo(ax, cx):
    lab = str(ax) + " - " + str(cx)
    plt.plot([foo(i, ax, cx) for i in range(T)], label=lab)
    plt.grid()
    plt.legend(loc='best')
```

```python
draw_foo(.1, 1 / 2)
draw_foo(.05, 1 / 2)
draw_foo(.1, 1 / 3)
draw_foo(.19, 1 / 3)
draw_foo(.3, 1 / 4)
```

```python
# Arms definition
def f1(x, horizon):
    # we want this arm to be optimum one, so it must reach the value 1
    lim = .1 * pow(horizon, 1 / 2)
    value = .1 * pow(x, 1 / 2)
    out = value/lim
    return out


def f2(x, horizon):
    # lim = .05 * pow(horizon, 1 / 2)
    lim = 1.1
    value = .05 * pow(x, 1 / 2)
    out = value/lim
    return out


def f3(x, horizon):
    # lim = .1 * pow(horizon, 1 / 3)
    lim = 1.2
    value = .1 * pow(x, 1 / 3)
    out = value/lim
    return out


def f4(x, horizon):
    # lim = .19 * pow(horizon, 1 / 3)
    lim = 1.5
    value = .19 * pow(x, 1 / 3)
    out = value/lim
    return out


def f5(x, horizon):
    # lim = 0.3 * pow(horizon, 1 / 4)
    lim = 1.5
    value = 0.3 * pow(x, 1 / 4)
    out = value/lim
    return out
```

```python
actions = [f1, f2, f3, f4, f5]
budget = 300

plt.clf()

i = 1
for arm in actions:
    label = str(i)
    plt.plot([arm(j, budget) for j in range(budget)], label=label)
    i += 1

plt.grid()
plt.legend(loc="best")
```

```python
# let's see the profiles of the curves
def df1(x):
    # we want this arm to be optimum one, so it must reach the value 1
    lim = .1 * pow(300, 1 / 2)
    # value = .1 * pow(x, 1 / 2)
    value = .1*0.5*pow(x, -0.5)
    out = value/lim
    return out

def df2(x):
    lim = 1.1
    # value = .05 * pow(x, 1 / 2)
    value = .05*.5*pow(x, -0.5)
    out = value/lim
    return out


def df3(x):
    lim = 1.2
    # value = .1 * pow(x, 1 / 3)
    value = .1*(1/3)*pow(x, (1/3)-1)
    out = value/lim
    return out


def df4(x):
    lim = 1.5
    # value = .19 * pow(x, 1 / 3)
    value = .19*(1/3)*pow(x, (1/3)-1)
    out = value/lim
    return out

def df5(x):
    lim = 1.5
    # value = 0.3 * pow(x, 1 / 4)
    value = .3*.25*pow(x, (1/4)-1)
    out = value/lim
    return out

def df6(x):
    return pow(x, -1.1)

# dactions = [df1, df2, df3, df4, df5, df6]
dactions = [df5, df6]
budget = 300

plt.clf()

i = 1
for arm in dactions:
    label = str(i)
    plt.plot([arm(j) for j in range(1, budget)], label=label)
    i += 1

plt.grid()
plt.legend(loc="best")
```
